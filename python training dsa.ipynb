```python
#superisor stack concept
def find_supervisors(arr):
    n = len(arr)
    result = [-1] * n  # Initialize the result array with -1
    stack = []  # Stack to keep track of indices

    # Traverse the list from right to left
    for i in range(n-1, -1, -1):
        # Maintain the elements in decreasing order in the stack
        while stack and arr[stack[-1]] <= arr[i]:
            stack.pop()
        
        # If stack is not empty, the top element is the next greater element
        if stack:
            result[i] = arr[stack[-1]]
        
        # Push the current index onto the stack
        stack.append(i)
    
    return result

# Example usage
input_array = [3, 5, 2, 14, 5, 3, 7, 9, 4, 6, 9, 4, 2, 5, 3]
output = find_supervisors(input_array)
print(output)
```

```python
#supervisor number code sir
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return None

    def top(self):
        if not self.is_empty():
            return self.items[-1]
        return None

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

# Given list
L = [3, 5, 2, 14, 5, 7, 9, 4, 2, 5, 3]
# Output list initialized to zero
O = [0] * len(L)
# Create a Stack instance
S = Stack()

# Traverse the list from right to left
for i in range(len(L) - 1, -1, -1):
    # Maintain the elements in decreasing order in the stack
    while not S.is_empty() and S.top() <= L[i]:
        S.pop()

    # If stack is empty, no greater element to the right
    if S.is_empty():
        O[i] = -1
    else:
        O[i] = S.top()
    
    # Push the current element onto the stack
    S.push(L[i])

print(O)
```

```python
#linked list insertion and deeletion
class Node:
    def __init__(self, data):
        self.value = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_front(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def insert_after(self, prev_node, data):
        if prev_node is None:
            print("The given previous node must be in the LinkedList.")
            return
        new_node = Node(data)
        new_node.next = prev_node.next
        prev_node.next = new_node

    def delete_front(self):
        if self.head is None:
            print("The list is empty.")
            return
        self.head = self.head.next

    def delete_end(self):
        if self.head is None:
            print("The list is empty.")
            return
        if self.head.next is None:
            self.head = None
            return
        second_last = self.head
        while second_last.next.next:
            second_last = second_last.next
        second_last.next = None

    def delete_node(self, key):
        temp = self.head

        if temp is not None:
            if temp.value == key:
                self.head = temp.next
                temp = None
                return

        while temp is not None:
            if temp.value == key:
                break
            prev = temp
            temp = temp.next

        if temp is None:
            print("The value is not present in the list.")
            return

        prev.next = temp.next
        temp = None

    def print_list(self):
        temp = self.head
        while temp:
            print(temp.value, end=" -> ")
            temp = temp.next
        print("None")

def main():
    llist = LinkedList()

    while True:
        print("1. Insert at front")
        print("2. Insert at end")
        print("3. Insert after a value")
        print("4. Delete from front")
        print("5. Delete from end")
        print("6. Delete a value")
        print("7. Print list")
        print("8. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            data = int(input("Enter value to insert at front: "))
            llist.insert_front(data)
        elif choice == '2':
            data = int(input("Enter value to insert at end: "))
            llist.insert_end(data)
        elif choice == '3':
            value = int(input("Enter value to insert after: "))
            data = int(input("Enter value to insert: "))
            current = llist.head
            while current and current.value != value:
                current = current.next
            if current is None:
                print("The given value is not in the list.")
            else:
                llist.insert_after(current, data)
        elif choice == '4':
            llist.delete_front()
        elif choice == '5':
            llist.delete_end()
        elif choice == '6':
            data = int(input("Enter value to delete: "))
            llist.delete_node(data)
        elif choice == '7':
            llist.print_list()
        elif choice == '8':
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
#adding two operators without using functions
def add_strings(num1, num2):
    # Ensure num1 is the longer string
    if len(num1) < len(num2):
        num1, num2 = num2, num1

    # Reverse the strings to start addition from the least significant digit
    num1 = num1[::-1]
    num2 = num2[::-1]

    carry = 0
    result = []

    # Loop through each digit and add
    for i in range(len(num1)):
        digit1 = int(num1[i])
        digit2 = int(num2[i]) if i < len(num2) else 0

        total = digit1 + digit2 + carry
        carry = total // 10
        result.append(str(total % 10))

    # If there's a carry left at the end, add it to the result
    if carry:
        result.append(str(carry))

    # Reverse the result to get the final sum and join the digits into a string
    return ''.join(result[::-1])

# Taking input from the user as strings
num1 = input("Enter the first number: ")
num2 = input("Enter the second number: ")

# Adding the numbers
sum_result = add_strings(num1, num2)

# Printing the result
print("The sum is:", sum_result)
```

```python
def evaluate_expression(expression):
    # Initialize variables
    num1 = ''
    num2 = ''
    operator = None
    found_operator = False

    # Iterate through each character in the expression
    for char in expression:
        if char in '+-*/':
            operator = char
            found_operator = True
        elif found_operator:
            num2 += char
        else:
            num1 += char

    # Convert string numbers to integers
    num1 = float(num1)
    num2 = float(num2)

    # Perform the arithmetic operation
    if operator == '+':
        return num1 + num2
    elif operator == '-':
        return num1 - num2
    elif operator == '*':
        return num1 * num2
    elif operator == '/':
        return num1 // num2  # Integer division

# Given input
a = 10 + 10  # Direct addition
b = "10+10"  # String input

# Evaluate the string expression
b_evaluated = evaluate_expression(b)

# Print the results
print(a, b_evaluated)
```

```python
# Input the elements,bubble sort 
L = list(map(int, input("Enter the elements separated by space: ").split()))

# Calculate the length of the list
n = len(L)

# Bubble sort algorithm
for j in range(n):
    # Last j elements are already sorted, so we don't need to check them
    for i in range(0, n - 1 - j):
        if L[i] > L[i + 1]:
            # Swap if the current element is greater than the next one
            L[i], L[i + 1] = L[i + 1], L[i]

# Print the sorted list
print("Sorted list:", L)
```

```python
# Python program for implementation of Selection
# Sort
A = [64, 25, 12, 22, 11]

# Traverse through all array elements
for i in range(len(A)-1):
    
    # Find the minimum element in remaining 
    # unsorted array
    min_idx = i
    for j in range(i+1, len(A)):
        if A[min_idx] > A[j]:
            min_idx = j
            
    # Swap the found minimum element with 
    # the first element        
    A[i], A[min_idx] = A[min_idx], A[i]

# Driver code to test above
print ("Sorted array")
for i in range(len(A)):
    print(A[i],end=" ") 
```

```python
# Function for nth Fibonacci number

def Fibonacci(n):
	if n<= 0:
		print("Incorrect input")
	# First Fibonacci number is 0
	elif n == 1:
		return 1
	# Second Fibonacci number is 1
	elif n == 2:
		return 1
	else:
		return Fibonacci(n-1)+Fibonacci(n-2)

# Driver Program

print(Fibonacci(10))
```

```python
#quick sort
def partition(L, Low, High):
    P = L[High]
    j = Low - 1
    for i in range(Low, High):
        if L[i] <= P:
            j += 1
            L[i], L[j] = L[j], L[i]
    j += 1
    L[j], L[High] = L[High], L[j]
    return j

def quick_sort(L, Low, High):
    if Low < High:
        pi = partition(L, Low, High)
        quick_sort(L, Low, pi - 1)
        quick_sort(L, pi + 1, High)
    return

if __name__ == "__main__":
    L = list(map(int, input("Enter the elements separated by space: ").split()))
    quick_sort(L, 0, len(L) - 1)
    print("Sorted Array:", L)
```

```python
#merge sort
def merge(array, left, mid, right):
    # Create temp arrays for left and right subarrays
    leftArray = array[left:mid + 1]
    rightArray = array[mid + 1:right + 1]

    i = j = 0
    k = left

    # Merge the temp arrays back into array[left..right]
    while i < len(leftArray) and j < len(rightArray):
        if leftArray[i] <= rightArray[j]:
            array[k] = leftArray[i]
            i += 1
        else:
            array[k] = rightArray[j]
            j += 1
        k += 1

    # Copy the remaining elements of leftArray[], if any
    while i < len(leftArray):
        array[k] = leftArray[i]
        i += 1
        k += 1

    # Copy the remaining elements of rightArray[], if any
    while j < len(rightArray):
        array[k] = rightArray[j]
        j += 1
        k += 1

def mergeSort(array, left, right):
    if left < right:
        mid = (left + right) // 2
        mergeSort(array, left, mid)
        mergeSort(array, mid + 1, right)
        merge(array, left, mid, right)

if __name__ == "__main__":
    arr = [12, 11, 13, 5, 6, 7]
    mergeSort(arr, 0, len(arr) - 1)
    print("Sorted array is:", arr)
```
